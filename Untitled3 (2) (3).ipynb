1.Uninformed Search Techniques: -BFS -DFS -IDS

```python
# Breadth First Search

L = [
    ('a', ['b', 'c']),
    ('b', ['d', 'e']),
    ('c', ['f']),
    ('d', ['g']),
    ('e', []),
    ('f', ['h']),
    ('g', []),
    ('h', ['i']),
    ('i', [])
]

graph = {node: neighbors for node, neighbors in L}

start = 'a'
goal = 'i'
queue = [[start]]
visited = []

while len(queue) > 0:
    path = queue.pop(0)
    node = path[-1]

    if node in visited:
        continue

    visited.append(node)

    if node == goal:
        print(path)
        break

    for neighbor in graph[node]:
        new_path = path + [neighbor]
        queue.append(new_path)
```

```python
#Depth First Search


L = [
    ('a', ['b', 'c']),
    ('b', ['d', 'e']),
    ('c', ['f']),
    ('d', ['g']),
    ('e', []),
    ('f', ['h']),
    ('g', []),
    ('h', ['i']),
    ('i', [])
]

graph = {node: neighbors for node, neighbors in L}

start = 'a'
goal = 'i'
stack = [[start]]
visited = []

while len(stack) > 0:
    path = stack.pop()
    node = path[-1]

    if node in visited:
        continue

    visited.append(node)

    if node == goal:
        print(path)
        break

    for neighbor in graph[node]:
        new_path = path + [neighbor]
        stack.append(new_path)
```

```python
#Iterative Deepening Search


L = [
    ('x', ['y', 'z']),
    ('y', ['a', 'b']),
    ('z', ['c']),
    ('a', ['d']),
    ('b', []),
    ('c', ['e']),
    ('d', ['f']),
    ('e', []),
    ('f', ['g']),
    ('g', [])
]

graph = {node: neighbors for node, neighbors in L}

start = 'x'
goal = 'g'
depth_limit = 5
found = False

for limit in range(depth_limit + 1):
    stack = [[start]]
    visited = []

    while len(stack) > 0:
        path = stack.pop()
        node = path[-1]

        if node in visited:
            continue

        visited.append(node)

        if node == goal:
            print(path)
            found = True
            break

        if len(path) - 1 < limit:
            for neighbor in graph[node]:
                new_path = path + [neighbor]
                stack.append(new_path)

    if found:
        break
```

2.Informed Search Techniques: -A* search technique -Best first search

```python
#A* Search :


L = [
    ('a', ['b', 'c'], 8),
    ('b', ['d', 'e'], 5),
    ('c', ['f'], 6),
    ('d', ['g'], 4),
    ('e', [], 3),
    ('f', ['h'], 2),
    ('g', [], 1),
    ('h', [], 0)
]

graph = {node: neighbors for node, neighbors, _ in L}
heuristics = {node: h for node, _, h in L}

start = 'a'
goal = 'h'
open_list = [[start, 0 + heuristics[start]]]
visited = []

while len(open_list) > 0:
    open_list.sort(key=lambda x: x[1])
    path = open_list.pop(0)
    node = path[0][-1]

    if node in visited:
        continue

    visited.append(node)

    if node == goal:
        print(path[0])
        break

    for neighbor in graph[node]:
        g_cost = len(path[0])
        f_cost = g_cost + heuristics[neighbor]
        new_path = path[0] + [neighbor]
        open_list.append([new_path, f_cost])
```

```python
#Best-First Search:

nodes = [
    ('i', ['j', 'k'], 7),
    ('j', ['m', 'n'], 6),
    ('k', ['o'], 5),
    ('m', ['p', 'q'], 4),
    ('n', ['r'], 3),
    ('o', ['s', 't'], 2),
    ('p', [], 6),
    ('q', [], 4),
    ('r', [], 5),
    ('s', [], 2),
    ('t', [], 0)
]

graph = {node: neighbors for node, neighbors, _ in nodes}
heuristic_values = {node: h for node, _, h in nodes}

initial_node = 'i'
target_node = 't'
open_set = [[initial_node, heuristic_values[initial_node]]]
explored = []

while len(open_set) > 0:
    open_set.sort(key=lambda x: x[1])  # Sort by heuristic value
    current_path = open_set.pop(0)[0]
    current_node = current_path[-1]

    if current_node in explored:
        continue

    explored.append(current_node)

    if current_node == target_node:
        print(current_path)
        break

    for neighbor in graph[current_node]:
        new_path = current_path + neighbor
        open_set.append([new_path, heuristic_values[neighbor]])
```

3. Adversal Search Techniques:
-(min-max)
-(alpha-beta)

```python
#min-max technique


game_tree = {
    'value': 0,
    'choices': [
        {'value': 0, 'choices': [
            {'value': 20, 'choices': []},
            {'value': 40, 'choices': []}
        ]},
        {'value': 0, 'choices': [
            {'value': 30, 'choices': []},
            {'value': 60, 'choices': []}
        ]}
    ]
}

max_depth = 3
is_maximizing = False

stack = [(game_tree, max_depth, is_maximizing)]
optimal_value = float('-inf') if is_maximizing else float('inf')

while stack:
    node, depth, is_max = stack.pop()

    if depth == 0 or not node['choices']:
        if is_maximizing:
            optimal_value = max(optimal_value, node['value'])
        else:
            optimal_value = min(optimal_value, node['value'])
    else:
        for choice in node['choices']:
            stack.append((choice, depth - 1, not is_maximizing))

print("Optimal value for maximizing player:", optimal_value)
```

Optimal score for maximizing player: 50

```python
#Alpha-Beta technique


graph = {
    'cost': 0,
    'children': [
        {'cost': 0, 'children': [
            {'cost': 4, 'children': []},
            {'cost': 7, 'children': []}
        ]},
        {'cost': 0, 'children': [
            {'cost': 2, 'children': []},
            {'cost': 5, 'children': []}
        ]}
    ]
}

depth = 2
is_minimizing = True
alpha = float('-inf')
beta = float('inf')

stack = [(graph, depth, is_minimizing, alpha, beta)]

best_cost = float('inf') if is_minimizing else float('-inf')

while stack:
    node, current_depth, is_min, alpha, beta = stack.pop()

    if current_depth == 0 or not node['children']:
        cost = node['cost']
        if is_minimizing:
            best_cost = min(best_cost, cost)
        else:
            best_cost = max(best_cost, cost)
    else:
        if is_min:
            min_eval = float('inf')
            for child in node['children']:
                stack.append((node, current_depth, is_min, alpha, beta))
                stack.append((child, current_depth - 1, False, alpha, beta))
                if min_eval <= alpha:
                    break
                min_eval = min(min_eval, best_cost)
                beta = min(beta, min_eval)
            best_cost = min_eval
        else:
            max_eval = float('-inf')
            for child in node['children']:
                stack.append((node, current_depth, is_min, alpha, beta))
                stack.append((child, current_depth - 1, True, alpha, beta))
                if max_eval >= beta:
                    break
                max_eval = max(max_eval, best_cost)
                alpha = max(alpha, max_eval)
            best_cost = max_eval

print("Optimal cost for minimizing player:", best_cost)
```
